## Rating Features
- There are two ways of scoring
    - Grade according to Quantiles: 'population_density','offence_count_scaled', 'percentage_rent'
    - Custom Scoring: 'num_stations','num_schools','num_hospitals'
- Sum all the score calculated to get final score 

```python
import pandas as pd
suburb_df = pd.read_csv("../data/curated/suburb_final.csv")
suburb_df
```

```python
# add new cloumn 'percentage_rent' calculate the  proportion of rental price in 2022_income
suburb_df['percentage_rent'] = 52*suburb_df['rental_price']/suburb_df['2022_income']
```

```python
# function to score quantile

'''
function for quantile scoring
- df is the dataframe used
- column is the feature name we are going to grade
- score is the new column storage score we generated for this feature
'''

def quantile_score(df,column,score):
    df[score] = 0
    df.loc[df[column] <=df[column].describe()[4],score] = 4
    df.loc[(df[column].describe()[4] < df[column])&( df[column]<= df[column].describe()[5]),score] = 3
    df.loc[(df[column].describe()[5] < df[column]) & (df[column] <= df[column].describe()[6]),score] = 2
    df.loc[df[column].describe()[6] < df[column],score] = 1

def quantile_score2(df,column,score):
    df[score] = 0
    df.loc[df[column] <=df[column].describe()[4],score] = 4
    df.loc[(df[column].describe()[4] < df[column])&( df[column]<= df[column].describe()[5]),score] = 2
    df.loc[(df[column].describe()[5] < df[column]) & (df[column] <= df[column].describe()[6]),score] = 1
    df.loc[df[column].describe()[6] < df[column],score] = 3
    
# scoring for population density
quantile_score2(suburb_df,'population_density','density_score')
# scoring for percentage_rent
quantile_score(suburb_df,'percentage_rent','percentage_rent_score')
# scoring for offence_count_scaled
quantile_score(suburb_df,'offence_count_scaled','offence_count_scaled_score')

suburb_df
```

```python
# Custom rating
suburb_df = suburb_df.fillna(0)
suburb_df.iloc[:,8:11] = suburb_df.iloc[:,8:11].values.astype(int)

'''
function for custom scoring standard, set different levels of valves
- df is the dataframe used
- column is the feature name we are going to grade
- score is the new column storage score we generated for this feature
- range_1 to range_4 Indicate the value range of four levels
'''

def custom_score(df,column,score,range_1,range_2,range_3,range_4):
    df[score] = 0
    df.loc[df[column].isin(range_1),score] = 1
    df.loc[df[column].isin(range_2),score] = 2
    df.loc[df[column].isin(range_3),score] = 3
    df.loc[df[column].isin(range_4),score] = 4
    
# scoring for number of station
custom_score(suburb_df,'num_stations','num_stations_score',range(0,1),range(1,2),range(2,4),range(4,6))
# scoring for number of schools
custom_score(suburb_df,'num_schools','num_schools_score',range(0,4),range(4,8),range(8,12),range(12,15))
# scoring for number of hospitals
custom_score(suburb_df,'num_hospitals','num_hospitals_score',range(0,1),range(1,3),range(3,6),range(6,9))
suburb_df
```

```python
# sum all calculated score to get final score
suburb_df['final_score'] = suburb_df.iloc[:,12:].sum(axis=1, skipna=True)
suburb_df = suburb_df.sort_values(by='final_score',ascending=False)
suburb_df.drop(suburb_df.columns[12:18], axis=1, inplace=True)
suburb_df
```
