```python
import pandas as pd
import geopandas as gpd
```

```python
population = pd.read_csv('../data/raw/population.csv')
properties = pd.read_csv('../data/raw/properties_preporcessed_1.csv')
```

```python
population=population.iloc[:,1:]
population.head(10)
```

```python
properties=properties.iloc[:,1:]
properties.head(20)
```

```python
display(population.columns)
display(properties.columns)
```

```python
df1=properties.groupby(['SA2_CODE21','LOC_PID'])['name'].count()
df1=df1.reset_index()
df1=df1.rename(columns={'name':'intersection_count'})
df1
```

```python
df2=properties.groupby(['SA2_CODE21'])['name'].count()
df2=df2.reset_index()
df2=df2.rename(columns={'name':'SA2_count'})

df2
```

```python
df3=pd.merge(df1,df2)
df3['ratio']=df3['intersection_count']/df3['SA2_count']

df3.SA2_CODE21=df3.SA2_CODE21.astype('int')
df3
```

```python
df_pop=population[['SA2 code','ERP 2021']]
df_pop=df_pop.rename(columns={'SA2 code':'SA2_CODE21'})
df_pop
```

```python
df4=pd.merge(df3,df_pop)

df4['suburb_population']=df4['ratio']*df4['ERP 2021']
df4
```

```python
df5=df4.groupby(['LOC_PID'])['suburb_population'].sum()
df5=df5.reset_index()
df5
```
