```python
import openrouteservice as ors
import folium
import pandas as pd


client = ors.Client(key='5b3ce3597851110001cf6248a9d8b3b41f024c41875c577ac7ce21ee')
import folium
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import json
pd.options.mode.chained_assignment = None
import requests
with open('/root/generic-real-estate-consulting-project-group-6/data/raw/property_raw.json','r',encoding='utf8')as fp:
    json_data = json.load(fp)

properties = pd.DataFrame(json_data.values())
properties.dropna()
```

```python
# This block of code finds out the distance of the properties to CBD Melbourne
CBD_LOCATION = [-37.8178197, 144.9667445]
result = []
for i in range(0, 25):
    if(i < 51):
        coordinates = [list(reversed(CBD_LOCATION))] + [list(reversed(coord)) for coord in properties.iloc[300*i:300*(i+1),:]['coordinates'].tolist()]
    else:
        coordinates = [list(reversed(CBD_LOCATION))] + [list(reversed(coord)) for coord in properties.iloc[300*i:,:]['coordinates'].tolist()]
    [[float("{:.7f}".format(x)), float("{:.7f}".format(y))] for [x,y] in coordinates]
    print(f"Start {i}th request request size {len(coordinates)}")
    matrix = client.distance_matrix(
        locations=coordinates,
        destinations = [0],
        profile='driving-car',
        metrics=['distance'],
        validate=False,
    )
    print(f"end {i}th request")
    curr = [dist[0] for dist in matrix['distances'][1:]]
    result.append(curr)
```
