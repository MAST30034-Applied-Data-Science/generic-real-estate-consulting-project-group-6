```python
import folium
map_2 = folium.Map(location=[-144, 37], tiles="Stamen Terrain", zoom_start=11)

# refer to the folium documentations on more information on how to plot aggregated data.
t = folium.Choropleth(
    geo_data=geoJSON, # geoJSON 
    name='Traffic Volumn Geographical Visualization', # name of plot
    ##data= zone_traffic_full, # data source
    ##columns=['LocationID','sqrt_traffic'], # the columns required
    ##key_on='properties.LocationID', # this is from the geoJSON's properties
    fill_color='Reds', # color scheme
    nan_fill_color='green',
    legend_name='sqrt of weighted average traffic volume'
)

t.add_to(map_2)

map_2
```
