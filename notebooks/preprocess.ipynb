```python
import numpy as np
import pandas as pd
import json
pd.options.mode.chained_assignment = None
# Read the yellow taxis data
with open('/root/generic-real-estate-consulting-project-group-6/data/raw/property_raw.json','r',encoding='utf8')as fp:
    json_data = json.load(fp)

df = pd.DataFrame(json_data.values())
df.dropna()
```

```python
ps_code = df['name']
ps_code1 = list(ps_code)
add = list(ps_code)
for i in range(len(ps_code1)-1):
    ps_code1[i] = ps_code1[i][-4:]
for i in range(len(add)-1):
    add[i] = add[i][:-9]
    
df['postcode'] = ps_code1
df['address'] = add
df 
```

```python
cost = list(df['cost_text'])
for i in range(len(cost)-1):
    if type(cost[i][2]) == 'int':
        cost[i] = cost[i][1:5]
    else:
        cost[i] = 0
cost

```
