```python
import numpy as np
import pandas as pd
import json
pd.options.mode.chained_assignment = None
# Read the yellow taxis data
with open('../data/raw/property_raw.json','r',encoding='utf8')as fp:
    json_data = json.load(fp)

df = pd.DataFrame(json_data.values())
df.dropna()
```

```python
ps_code = df['name']
ps_code1 = list(ps_code)
add = list(ps_code)
for i in range(len(ps_code1)-1):
    ps_code1[i] = ps_code1[i][-4:]
for i in range(len(add)-1):
    add[i] = add[i][:-9]
    
df['postcode'] = ps_code1
df['address'] = add
df 
```

```python
cost = list(df['cost_text'])
for i in range(len(cost)-1):
    if type(cost[i][2]) == 'int':
        cost[i] = cost[i][1:5]
    else:
        cost[i] = 0
cost

```

```python
import folium
from sklearn.model_selection import train_test_split
json0, json1 = train_test_split(df, test_size=0.1, random_state=0)
json1
```

```python
coordinates = json1["coordinates"]
mapit = folium.Map( location=[-39, 145], zoom_start=6)
for coord in coordinates:
    folium.Marker( location=[ coord[0], coord[1] ], fill_color='#43d9de', radius=8 ).add_to( mapit )
```
