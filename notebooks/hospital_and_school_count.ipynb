```python
# import libraries
# read in the hospital data

import pandas as pd
import numpy as np
import seaborn as sns

df_hospital =  pd.read_csv('../generic-real-estate-consulting-project-group-6/data/raw/MainHospitalList.csv',header=0)
df_hospital
```

```python
# Count Victoria ten suburbs with the most hospitals
 

df_suburb = df_hospital.Suburb
df_count_hospital = df_suburb.value_counts()
hospital_data = df_count_hospital[0:10]
hospital_data.head(10)
```

```python
# Visualize the top ten suburbs with the most hospitals in Victoria

sns.barplot(x=hospital_data.values, y=hospital_data.keys())
```

```python
# Read in the school data

df_school =  pd.read_csv('../generic-real-estate-consulting-project-group-6/data/raw/dv309_schoollocations2021.csv', sep=",", encoding='cp1252')
df_school
```

```python
# Count Victoria ten suburbs with the most schools
# As there are 3 suburbs have 10 schools, the data will choose top 11

df_Postal_Town = df_school.Postal_Town
df_count_school = df_Postal_Town.value_counts()
school_data = df_count_school[0:11]
school_data.head(11)
```

```python
#Visualize the top ten suburbs with the most schools in Victoria

sns.barplot(x=school_data.values, y=school_data.keys())
```
